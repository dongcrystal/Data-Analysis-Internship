```python
import pandas as pd
import numpy as np
import folium
```

```python
from google.colab import files
road_file = files.upload()
```

```python
df = pd.read_csv('The latitude and longitude of popular tourist attractions in Jeju.csv')
df.tail()
```

```python
latitude = df['latitude']
latitude = [float(i) for i in latitude]
longtitude = df['longitude']
longtitude = [float(i) for i in longtitude]
```

```python
m = folium.Map(location = [33.3617,126.5292],title = 'jeju',zoom_start = 11)
for i in range(0,len(df)):
    latitude2 = latitude[i]
    longtitude2 = longtitude[i]
    radius2 = float(df.iloc[i]['외지인 검색 수']/16000)
    location = [latitude2,longtitude2]
    folium.CircleMarker(location,radius = radius2,color="#000099",fill_color="000099").add_to(m)
m.save('./jeju.html')
```
