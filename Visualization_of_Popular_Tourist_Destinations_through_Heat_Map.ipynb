```python
import pandas as pd
import folium
from folium.plugins import HeatMap
```

```python
from google.colab import files
road_file = files.upload()
```

```python
df = pd.read_csv('The latitude and longitude of popular tourist attractions in Jeju.csv')
df
```

```python
m = folium.Map(location = [33.36,126.52], tiles = 'openstreetmap', zoom_start = 10)
HeatMap(
    data = df[['latitude','longitude']],
    radius = 20).add_to(m)
m.save('./map1.html')
```
